```python
import zipfile
import pandas as pd
```

```python
# Path to the zip file
zip_file_path = 'export_41.zip'

# Extract the CSV files from the zip archive
with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
    csv_files = [file for file in zip_file.namelist() if file.endswith('.csv')]
    zip_file.extractall()
```

```python
# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate through the extracted CSV files and merge them into a single DataFrame
for csv_file in csv_files:
    data = pd.read_csv(csv_file, delimiter=';')  # Read each CSV file
    merged_data = pd.concat([merged_data, data])  # Concatenate the data to the merged DataFrame

# Write the merged data to a new CSV file
merged_data.to_csv('merged_doughnut.csv', index=False)

print("CSV files from the zip archive have been successfully merged into 'merged.csv'.")
```
